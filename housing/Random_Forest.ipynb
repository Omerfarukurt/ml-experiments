```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('brain_tumor_dataset.csv')
```

```python
data.head()
```

```python
df=data.copy()
```

```python
df.Tumor_Type.value_counts()
```

```python
df['Tumor_Type']=df['Tumor_Type'].map({'Malignant':0,'Benign':1})
```

```python
df.Location.value_counts()
```

```python
cat_cols=df.select_dtypes(include='object').columns
```

```python
df.Stage.value_counts()
```

```python
df['Stage']=df['Stage'].map({'I':1,'II':2,'III':3,'IV':4})
```

```python
X=df.drop('Tumor_Type',axis=1)
```

```python
y=df['Tumor_Type']
```

```python
X=pd.get_dummies(data=X,columns=cat_cols)
```

```python
X.head()
```

```python
bool_column=X.select_dtypes(include=["bool"]).columns
X[bool_column]=X[bool_column].astype(int)
```

```python
X.head()
```

```python
X.drop('Patient_ID',axis=1,inplace=True)
```

```python
cols=X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
X=pd.DataFrame(data=X,columns=cols)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
fit_clf=RandomForestClassifier()
```

```python
np.random.seed(42)
```

```python
param={'max_depth':[2,3,4,5,6,8],
      'max_features':['auto','sqrt','log2',None],
      'n_estimators':[50,100,150,200,250,300,350,400],
       'criterion':['gini','entropy']
      }
```

```python
rf_cv=GridSearchCV(fit_clf,
                  cv=2,
                  param_grid=param,
                  n_jobs=4)
```

```python
rf_cv.fit(X_train,y_train)
```

```python
rf_cv.best_params_
```

```python
fit_clf.set_params(criterion='entropy',
                  max_depth=1,
                  max_features='sqrt',
                  n_estimators=250)
```

```python
fit_clf.fit(X_train,y_train)
```

```python
y_pred=fit_clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,roc_auc_score
```

```python
roc_auc_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
