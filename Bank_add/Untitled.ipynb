```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,roc_auc_score,confusion_matrix
```

```python
data.head()
encode=LabelEncoder()
```

```python
for i in cat_cols:
    data[i]=encode.fit_transform(data[i])
    
```

```python
X=data.drop('y',axis=1)
y=data['y']
```

```python
col_names=X.columns
```

```python
scaler=MinMaxScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
X=pd.DataFrame(data=X,columns=col_names)
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
tree_clf=DecisionTreeClassifier(criterion='entropy',max_depth=7,random_state=20)
```

```python
tree_clf.fit(X_train,y_train)
```

```python
y_pred=tree_clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
data['y'].value_counts()
```

```python
tree_clf.feature_importances_
```

```python
important=tree_clf.feature_importances_
names=X.columns
```

```python
imp=sorted(zip(important,names),reverse=True)
```

```python
for important,features in imp:
    print(f'{features} Ã¶nem derecesi:{important}')
```

```python
df=data.copy()
onemsiz=[]
```
